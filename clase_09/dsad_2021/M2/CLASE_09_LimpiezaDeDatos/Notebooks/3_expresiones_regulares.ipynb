## Intro

[volver a TOC](#section_toc)

Una expresin regular es una secuencia de caracteres que define un **patrn** de bsqueda de texto. 

Las expresiones regulares constituyen un lenguaje muy flexible que sirve para identificar y extraer informacin de un cuerpo de texto.

Una sola expresin, comnmente llamada regex, es una cadena de caracteres construida de acuerdo a la sintaxis del lenguaje de expresiones regulares. 

El mdulo `re` de Python es responsable de aplicar expresiones regulares a cadenas de caracteres.

Las funciones del mdulo `re` se agupan en tres categoras:

* Split: divide una cadena de caracteres, usando el patrn definido por la expresin regular como separador.

* Pattern maching: extrae, de una cadena de caracteres, las subcadenas definidas por el patrn de la expresin regular

* Sustitucin: reemplaza la cadena de caracteres definida por la expresin regular por otra


<a id="section_sintaxis"></a> 
### Sintaxis en expresiones regulares en Python

[volver a TOC](#section_toc)

#### Metacaracteres

    .  smbolo que indica cualquier caracter con excepcin de nueva lnea (\n)
    ^  smbolo que indica comienzo
    $  smbolo que indica fin
    \  smbolo que escapa caracteres reservados
    |  o lgico 
    \d smbolo que indica cualquier dgito del 0 al 9
    \w smbolo que indica cualquier caracter alfanumrico (A-Z, a-z, 0-9 y _)
    \s smbolo que indica cualquier espacio en blanco (espacio, tabulado, nueva lnea, etc.)
    \D smbolo que indica cualquier caracter que no sea dgito
    \W smbolo que indica cualquier caracter que no sea alfanumrico
    \S smbolo que indica cualquier caracter que no sea espacio en blanco

        
#### Cuantificadores

    *  smbolo que indica cero o ms ocurrencias
    +  smbolo que indica una o ms ocurrencias
    ?  smbolo que indica optativo (puede estar o no)
    {m}  donde m es un nmero entero, indica exactamente m repeticiones 
    {m,n}  donde m y n son nmeros enteros, indica al menos m repeticiones y como mximo n repeticiones

#### Opciones 

    [abc]  indica un caracter perteneciente al conjunto de valores posibles especificado entre corchetes
    [a-z]  indica un caracter perteneciente al intervalo de valores posibles especificado entre corchetes

#### Grupos 
    ( )  define un grupo
    (?P<group_name> ) define un grupo etiquetado
    
#### Ejemplos de sintaxis:

`\d+` encuentra nmeros de un dgito o ms

`.*` encuentra cadenas de caracteres de cualquier longitud, incluso vacas.

`\w{2,6}` encuentra un conjunto de caracteres alfanumricos con una longitud que va de 2 a 6 caracteres

`[a-zA-Z]` encuentra lo mismo que `[a-zA-Z]{1}` que es cualquier caracter entre a y z minscula o A y Z mayscula

`[a-zA-Z]+` encuentra cadenas de caracteres con letras entre a y z y A y Z de longitud al menos 1

`(\d\d\d\d)` encuentra lo mismo que `\d{4}` que son cadenas de cuatro dgitos

`(?P<num>\d\d\d\d)` encuentra  cadenas de cuatro dgitos y etiqueta a ese grupo como "num"

<a id="section_split"></a> 
### Split

[volver a TOC](#section_toc)

Dividimos una cadena de caracteres (la variable texto), usando el patrn definido en la expresin regular como separador.

Obtenemos como resultado la lista de partes que componen la variable texto.

De link de referencia de sintaxis vemos que

*\s: Returns a match where the string contains a white space character*

```python
import re

patron = "\s+"

regex = re.compile(patron, flags = re.IGNORECASE)

texto = "foo      bar \t bas \tqux"
regex.split(texto)
```

<a id="section_pattern_matching"></a> 
### Pattern matching

[volver a TOC](#section_toc)

Si buscamos slo el primer match del patrn de la expresin regular:

```python
regex.search(texto)
```

match slo busca al comienzo del string:

```python
result = regex.match(texto)
print(result)
```

```python
texto_2 = '   ' + texto
result = regex.match(texto_2)
print(result)
```

<a id="section_ejemplos"></a> 
### Ejemplos

[volver a TOC](#section_toc)



Usando en dataset de porpiedades de Melbourne, el campo Address, vamos a extraer el nmero y el nombre de la calle sin el sufijo 'St'


Leemos los datos:    

```python
import pandas as pd 

# local
data_location = "../Data/melb_data.csv"
# colab
# data_location = ""

data = pd.read_csv(data_location)

data.head(5)
```

La expresin regular para el nmero de la direccin es, cualquier dgito (\d) una o ms veces (+)

Como queremos extraerlo, vamos a usar alguna de las funciones de pattern matching, en este caso `re.findall`

Y como tenemos que evaluarlo sobre cada elemento de la series `data.Address` vamos a usar `apply`

```python
patron_num = "\d+"
patron_num_regex = re.compile(patron_num)
```

```python
address_series = data.Address
resultado = address_series.apply(lambda x: patron_num_regex.findall(x))
print(type(resultado))
print(resultado)
```

Obsevemos que:

* El resultado es una instancia de Series.

* Cada uno de sus elementos es una lista de longitud igual a la cantidad de nmeros que haya en el valor del campo Address de cada registro (porque findall devuelve todos los match que encuentra)


Si usamos search obtenemos slo el primero:

```python
address_series = data.Address
resultado = address_series.apply(lambda x: patron_num_regex.search(x))
print(type(resultado))
print(resultado)
```

Obsevemos que:

* El resultado es una instancia de Series.

* Cada uno de sus elementos es un objeto de tipo Match, que corresponde al primer nmero encontrado para cada registro en ese campo
https://docs.python.org/3/library/re.html#match-objects

Para extraer el valor del nmero a partir del objeto Match podemos usar `group` o `[ ]`:

```python
numeros_match = resultado.apply(lambda x: x[0])
numeros_match
```

```python
numeros_match = resultado.apply(lambda x: x.group(0))
numeros_match
```

Ahora busquemos el nombre de la calle sin el sufijo. Los sufijos pueden ser St, La, Cr, Dr.

Vamos a construir un patrn que identifique 

1)  el nmero de calle `(?P<numero_calle>\d+[a-z]*)` como uno o ms dgitos (\d+) seguidos opcionalemnet (*) por un caracter en minscula. Definimos el nombre del grupo con (?P<nombre_del_gripo> ...)
  
2)  seguido por un espacio (\s)

3) seguido por el nombre de la calle `(?P<nombre_calle>.+)` cualquier cadena de caracteres de longitud al menos 1

4) seguido por un sufijo `(?P<sufijo>\sSt|La|Cr|Dr)` que es un espacio seguido por la cadena de caracteres "St"  "La"  "Cr"  "Dr"

```python

pattern_nombre = "(?P<numero_calle>\d+[a-z]*)\s(?P<nombre_calle>.+)(?P<sufijo>\sSt|La|Cr|Dr)"

pattern_nombre_regex = re.compile(pattern_nombre)
address_series = data.Address
resultado_nombres = address_series.apply(lambda x: pattern_nombre_regex.search(x))
#print(type(resultado))
print(resultado_nombres)
```

Para extraer el nombre de la calle a partir del objeto Match podemos usar `group` sabiendo que el nombre del grupo es 'nombre_calle':

```python
calles_match = resultado_nombres.apply(lambda x: x if x is None else x.group('nombre_calle'))
calles_match
```

Ahora queremos dividir cada el campo Address de cada registro en tres partes: nmero, nombre y sufijo.
    
Para eso vamos a usar la misma expresin regular del punto anterior y el mtodo `split`

```python
pattern_address = "(?P<numero_calle>\d+[a-z]*)\s(?P<nombre_calle>.+)(?P<sufijo>\sSt|La|Cr|Dr)"

pattern_address_regex = re.compile(pattern_address)
address_series = data.Address
resultado_nombres = address_series.apply(lambda x: pattern_address_regex.split(x))
#print(type(resultado))
print(resultado_nombres)
```

Para terminar vamos a reemplazar el sufijo de cada registro por un string vaco


```python
pattern_suffix = "\sSt|La|Cr|Dr"

cadena_reemplazo = ""

pattern_suffix_regex = re.compile(pattern_suffix)
address_series = data.Address
resultado_reemplazo = address_series.apply(lambda x: pattern_suffix_regex.sub(cadena_reemplazo, x))
print(resultado_reemplazo)
```
